# Matplotlib

```python
import matplotlib.pyplot as plt
from scipy.stats import norm, t
import numpy as np
```

```python
mean = 0
sd = 1
x = np.arange(-3.5,3.5,0.025)
y = norm.pdf(x,loc=mean,scale=sd)
```

## Plot and shade a specific portion of the curve
The following features can be drawn:
1. Area shading
1. Axis labels
1. Chart title
1. Legend

```python
confidence_interval = 0.90
prob_high = (1+confidence_interval)/2
prob_low = (1-confidence_interval)/2
x_higher = norm.ppf(prob_high,loc=mean,scale=sd)
x_lower = norm.ppf(prob_low,loc=mean,scale=sd)
plt.plot(x,y,'-')
plt.fill_between(x[x>=x_lower],y[x>=x_lower],color='green')
plt.fill_between(x[x>=x_higher],y[x>=x_higher],color='white')
plt.legend(['PDF','CDF'], loc='upper right')
# add some vertical lines
plt.vlines(x_lower, ymin=0,ymax=max(y), colors='red')
plt.vlines(x_higher, ymin=0,ymax=max(y), colors='red')
plt.vlines(mean, ymin=0,ymax=max(y), colors='red')
plt.draw()
```
